```python
rank=pd.read_csv("../team_data/fifa_ranking-2024-06-20.csv")
rank.info()
```

```python
rank["rank_date"] = pd.to_datetime(rank["rank_date"], utc=True)
```

```python
rank=rank[rank["rank_date"].dt.year>=2022]
rank
```

```python
world_cup_2026_teams = [
    # Hosts
    "Canada", "Mexico", "USA",

    # AFC (Asia)
    "Australia", "IR Iran", "Japan", "Jordan", "Korea Republic",
    "Qatar", "Saudi Arabia", "Uzbekistan",

    # CAF (Africa)
    "Algeria", "Cabo Verde", "Côte d'Ivoire", "Egypt", "Ghana",
    "Morocco", "Senegal", "South Africa", "Tunisia",

    # CONCACAF (North & Central America, Caribbean)
    "Curacao", "Haiti", "Panama",

    # CONMEBOL (South America)
    "Argentina", "Brazil", "Colombia", "Ecuador", "Paraguay", "Uruguay",

    # OFC (Oceania)
    "New Zealand",

    # UEFA (Europe)
    "Austria", "Belgium", "Bosnia and Herzegovina", "Croatia", "Czechia",
    "England", "France", "Germany", "Netherlands", "Norway",
    "Portugal", "Scotland", "Spain", "Sweden", "Switzerland", "Turkey",

    # Play-Off Tournament winners
    "Bolivia", "Congo DR", "Iraq"
]
print(len(world_cup_2026_teams))
rank=rank[rank["country_full"].isin(world_cup_2026_teams)]
rank
```

```python
rank["confederation"]=rank.groupby("country_full")["confederation"].transform("first")
rank
```

```python
rank=rank.reset_index()
rank=rank.drop(columns=["country_abrv","index"])
rank
```

```python
rank_date=np.unique(rank["rank_date"])
rank_date
```

```python
def give_rank_date(date):
    for d in rank_date:
        if date>d:
            return d
       
matches["date"]=pd.to_datetime(matches["date"],utc=True)
matches["rank_date"]=matches["date"].apply(give_rank_date)
matches
```

```python
spcial_teams={"IR Iran":"Iran","Turkey":"Turkiye","Korea Republic":"South Korea","Côte d'Ivoire":"Ivory Coast","Congo DR":"DR Congo","Cabo Verde":"Cape Verde","Curacao":"Curaçao","USA":"United States","Czech Republic":"Czechia"}

spcial = {v: k for k, v in spcial_teams.items()}
```

```python
rank["country_full"]=rank["country_full"].replace(spcial)
```

```python
rank
```

```python
matches = matches.merge(
    rank[[
        "country_full",
        "rank_date",
        "rank",
        "total_points",
        "rank_change",
        "previous_points",
        "confederation"
    ]],
    left_on=["home_team", "rank_date"],
    right_on=["country_full", "rank_date"],
    how="left"
)

matches = matches.rename(columns={
    "rank": "home_rank",
    "total_points": "home_total_points",
    "rank_change": "home_rank_change",
    "previous_points": "home_previous_points",
    "confederation": "home_conf"
}).drop(columns=["country_full"])
matches = matches.merge(
    rank[[
        "country_full",
        "rank_date",
        "rank",
        "total_points",
        "rank_change",
        "previous_points",
        "confederation"
    ]],
    left_on=["away_team", "rank_date"],
    right_on=["country_full", "rank_date"],
    how="left"
)

matches = matches.rename(columns={
    "rank": "away_rank",
    "total_points": "away_total_points",
    "rank_change": "away_rank_change",
    "previous_points": "away_previous_points",
    "confederation": "away_conf"
}).drop(columns=["country_full"])
```
