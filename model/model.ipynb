```python
rank["confederation"]=rank.groupby("country_full")["confederation"].transform("first")
rank
```

```python
rank=rank.reset_index()
rank=rank.drop(columns=["country_abrv","index"])
rank
```

```python
rank_date=np.unique(rank["rank_date"])
rank_date
```

```python
def give_rank_date(date):
    for d in rank_date:
        if date>d:
            return d
       
matches["date"]=pd.to_datetime(matches["date"],utc=True)
matches["rank_date"]=matches["date"].apply(give_rank_date)
matches
```

```python
spcial_teams = {
    "IR Iran": "Iran",
    "Turkey": "Turkiye",
    "Korea Republic": "South Korea",
    "Côte d'Ivoire": "Ivory Coast",
    "Congo DR": "DR Congo",
    "Cabo Verde": "Cape Verde",
    "Curacao": "Curaçao",
    "USA": "United States",
    "Czech Republic": "Czechia"
}

rank["country_full"] = rank["country_full"].replace(spcial_teams)
matches["home_team"] = matches["home_team"].replace(spcial_teams)
matches["away_team"] = matches["away_team"].replace(spcial_teams)
```

```python
rank
```

```python
rank["country_full"] = rank["country_full"].str.strip()
matches["home_team"] = matches["home_team"].str.strip()
matches["away_team"] = matches["away_team"].str.strip()
```

```python
matches = matches.merge(
    rank[[
        "country_full",
        "rank_date",
        "rank",
        "total_points",
        "rank_change",
        "previous_points",
        "confederation"
    ]],
    left_on=["home_team", "rank_date"],
    right_on=["country_full", "rank_date"],
    how="left"
)

matches = matches.rename(columns={
    "rank": "home_rank",
    "total_points": "home_total_points",
    "rank_change": "home_rank_change",
    "previous_points": "home_previous_points",
    "confederation": "home_conf"
}).drop(columns=["country_full"])
matches = matches.merge(
    rank[[
        "country_full",
        "rank_date",
        "rank",
        "total_points",
        "rank_change",
        "previous_points",
        "confederation"
    ]],
    left_on=["away_team", "rank_date"],
    right_on=["country_full", "rank_date"],
    how="left"
)

matches = matches.rename(columns={
    "rank": "away_rank",
    "total_points": "away_total_points",
    "rank_change": "away_rank_change",
    "previous_points": "away_previous_points",
    "confederation": "away_conf"
}).drop(columns=["country_full"])
```

```python
matches=matches.dropna()
matches.info()
```

```python
matches["hour"]=matches["date"].dt.hour
matches
```

```python
matches["rank_diff"]=matches["home_rank"]-matches["away_rank"]
matches["points_diff"]=matches["home_total_points"]-matches["away_total_points"]
matches["age_diff"]=matches["home_team_age"]-matches["away_team_age"]
```

```python
matches_train=matches[matches["date"].dt.year>2023]
matches_test=matches[matches["date"].dt.year<2023]
x_train=matches_train.drop(columns=["date","home_team_goals","rank_date","away_team_goals","home_rank","away_rank","home_team_age","away_team_age"])
y_train=matches_train[["home_team_goals","away_team_goals"]]

```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.pipeline import Pipeline

team_encoder=LabelEncoder()
conf_encoder=LabelEncoder()

team_encoder.fit(pd.concat([matches["home_team"], matches["away_team"]]))
conf_encoder.fit(pd.concat([matches["home_conf"], matches["away_conf"]]))
x_train["home_team"]=team_encoder.transform(x_train["home_team"])
x_train["away_team"]=team_encoder.transform(x_train["away_team"])
x_train["home_conf"]=conf_encoder.transform(x_train["home_conf"])
x_train["away_conf"]=conf_encoder.transform(x_train["away_conf"])
```

```python
from sklearn.linear_model import LinearRegression

linear_model=Pipeline([("Scaler",StandardScaler()),("Regresso",LinearRegression())])
linear_model.fit(x_train,y_train)
```
