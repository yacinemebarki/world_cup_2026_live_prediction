```python
import numpy as np
import pandas as pd
import os 
import re
```

## extract the data

```python
pattern1=r"matches"
matches=pd.DataFrame()
players=pd.DataFrame()
data="../data"
for dirc in os.listdir(data):
    path=os.path.join(data,dirc)
    if os.path.isdir(path):
        print(dirc)
        for file in os.listdir(path):
            match=re.search(pattern1,file)
            if match:
                target_file=os.path.join(path,file)
                matches2=pd.read_csv(target_file)
                if matches.empty:
                    matches=matches2
                else :
                    matches=pd.concat([matches,matches2])
            else :
                target_file=os.path.join(path,file)
                players2=pd.read_csv(target_file)
                if players.empty:
                    players=players2
                else :
                    players=pd.concat([players,players2])   
                 
matches                
                        
                  
            
        
```

```python
def get_birthyear(date):
    i=0
    year=""
    while date[i] != "/" and i<len(date):
        year=year+date[i]
        i+=1
    return int(year)
print(get_birthyear("1991/05/05"))
```

```python
players["birthday_GMT"]=players["birthday_GMT"].apply(get_birthyear)
```

```python
matches["date_GMT"]=pd.to_datetime(matches["date_GMT"],errors="coerce")
matches["season"]=matches["date_GMT"].dt.year
matches["hour"]=matches["date_GMT"].dt.hour
matches
```

```python
matches=matches.merge(
    players,
    left_on=["home_team_name","season"],
    right_on=["Current Club","season"],
    how="left"
)
matches.rename(columns={"age": "home_team_age"}, inplace=True)
matches.drop(columns=["Current Club"], inplace=True)
matches=matches.merge(
    players,
    left_on=["away_team_name","season"],
    right_on=["Current Club","season"],
    how="left"
)
matches.rename(columns={"age": "away_team_age"}, inplace=True)
matches.drop(columns=["Current Club"], inplace=True)
matches
```

```python
game_weak=["G1","G2","G3","R16","R8","R4","R3","R2"]
max_matches=[0,16,32,48,56,60,62,63,64]
matches["index"] = matches.groupby("season").cumcount() + 1

matches["game_week"] = pd.cut(
    matches["index"],
    bins=max_matches,
    labels=game_weak
)
matches

```
