```python
import numpy as np
import pandas as pd
import os 
import re
```

## extract the data

```python
pattern1=r"matches"
matches=pd.DataFrame()
players=pd.DataFrame()
data="../data"
for dirc in os.listdir(data):
    path=os.path.join(data,dirc)
    if os.path.isdir(path):
        print(dirc)
        for file in os.listdir(path):
            match=re.search(pattern1,file)
            if match:
                target_file=os.path.join(path,file)
                matches2=pd.read_csv(target_file)
                if matches.empty:
                    matches=matches2
                else :
                    matches=pd.concat([matches,matches2])
            else :
                target_file=os.path.join(path,file)
                players2=pd.read_csv(target_file)
                if players.empty:
                    players=players2
                else :
                    players=pd.concat([players,players2])   
                 
matches                
                        
                  
            
        
```

```python
def get_birthyear(date):
    i=0
    year=""
    while date[i] != "/" and i<len(date):
        year=year+date[i]
        i+=1
    return int(year)
print(get_birthyear("1991/05/05"))
```

```python
players["birthday_GMT"]=players["birthday_GMT"].apply(get_birthyear)
```
